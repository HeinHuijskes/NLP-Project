```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
parameter = 'rid'
genres = ['rap', 'rock', 'rb', 'country']
xs, ys = [], []
limits = [1000]
for limit in limits:
    file_name = f'output/{limit}/{parameter}/{"_".join(genres)}.csv'
    df = pd.read_csv(file_name)
    xs.append(df[parameter])
    ys.append(df['accuracy'])

fig, (ax1) = plt.subplots(1, 1)
fig.suptitle(f'{parameter}')
ax1.set_title('accuracy')
for i in range(len(xs)):
    ax1.plot(xs[i], ys[i])
fig.legend(limits)

print(df)

plt.show()
```

```python
parameter = 'cutoff'
genres = ['rap', 'rock', 'pop']
# genres = ['rap', 'rock', 'pop', 'rb', 'country']
xs, ys, zs = [], [], []
limits = [10000]
for limit in limits:
    file_name = f'output/{limit}/{parameter}/{"_".join(genres)}.csv'
    df = pd.read_csv(file_name)
    xs.append(df[parameter])
    ys.append(df['accuracy'])
    zs.append(df['features'])

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle(f'{parameter}')
ax1.set_title('accuracy')
ax2.set_title('features')
for i in range(len(xs)):
    ax1.plot(xs[i], ys[i])
    ax2.plot(xs[i], zs[i])
fig.legend(limits)

print(df)

plt.show()
```

```python
parameter = 'stem'

xs, ys = [], []
limits = [100, 1000]
for limit in limits:
    file_name = f'output/{limit}/{parameter}/{"_".join(genres)}.csv'
    df = pd.read_csv(file_name)
    xs.append(list(df[parameter]))
    ys.append(list(df['accuracy']))

fig, axs = plt.subplots(1, len(limits))
fig.suptitle(f'{parameter}')
for i, ax in enumerate(axs):
    ax.set_title(f'{limits[i]}')
    ax.bar(xs[i], ys[i])
plt.show()
```
