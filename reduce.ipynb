```python
import pandas as pd

df = pd.read_csv('data/song_lyrics_min.csv')
```

```python
df['tag'].unique()
```

```python
def make_data(genres, data, lim=-1):
    new_df = data.loc[data['tag'].isin(genres)]
    print(len(data), len(new_df))
    for genre in genres:
        print('Genre', genre, len(new_df.loc[new_df['tag'] == genre]))
    new_data = pd.DataFrame(columns=new_df.columns)
    for genre in genres:
        addition = new_df.loc[new_df['tag'] == genre][:lim]
        new_data = pd.concat([new_data, addition], ignore_index=True)
    print(len(new_data), len(new_df), len(data))
    if lim == -1:
        lim = 'all'
    new_data.to_csv(f'data/song_lyrics_reduced_{"_".join(genres)}_{lim}.csv')
    print(f'Succesfully wrote {genres} with size {len(new_data)} to file (lim={lim})')
```

```python
genres = ['rap', 'rock', 'pop']
lim = 1000
make_data(genres, df, lim)
```
