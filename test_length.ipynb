```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from nltk import FreqDist

from DataPrep import preprocess, remove_rare_words, remove_notes, remove_punctuation, normalize, tokenize
```

```python
genres = ['rap', 'rock', 'rb', 'country']
limit = 1000

cutoff = 10
stem = True
rid=20

df = pd.read_csv(f'data/song_lyrics_reduced_{"_".join(genres)}_{limit}.csv')
# documents = preprocess(df['lyrics'], debug=True, limit=cutoff, stem_words=stem, use_rid=rid)
noteless = remove_notes(df['lyrics'])
line_counts = []
for document in noteless:
    line_counts.append(len(document.split('\n')))
print(line_counts)

distribution = FreqDist(line_counts)
unique = list(distribution)
print(unique)
print(max(unique), min(unique), sum(unique)/len(unique), len(unique), len(line_counts))
```

```python
minimum = min(unique)
maximum = max(unique)
for bucket in range(1, 50):
    if bucket == 0:
        bucket = 1
    buckets = [0 for i in range(bucket)]
    for length in unique:
        index = int(((length - minimum) / (maximum - minimum) * bucket) // 1)
        buckets[min(index, bucket-1)] += 1
    print(bucket, [float(x) for x in np.quantile(line_counts, [0.25, 0.5, 0.75])], buckets)
```

```python
just_words = remove_punctuation(noteless)
normalized = normalize(just_words)
tokens = tokenize(normalized)

token_counts = []
for document in tokens:
    token_counts.append(len(document))
print(token_counts)

distribution_tokens = FreqDist(token_counts)
unique_token_counts = list(distribution_tokens)
print(unique_token_counts)
print(max(unique_token_counts), min(unique_token_counts), sum(unique_token_counts)/len(unique_token_counts), len(unique_token_counts), len(token_counts))
```

```python
minimum = min(unique_token_counts)
maximum = max(unique_token_counts)
for bucket in range(1, 50):
    buckets = [0 for i in range(bucket)]
    for length in unique_token_counts:
        index = int(((length - minimum) / (maximum - minimum) * bucket) // 1)
        buckets[min(index, bucket-1)] += 1
    print(bucket, [float(x) for x in np.quantile(buckets, [0.25, 0.5, 0.75, 1.0])], buckets)
```

```python
tokens_per_line = []
for document in range(len(noteless)):
    tokens_per_line.append(int(token_counts[document] // line_counts[document]))

dist_tpl = FreqDist(tokens_per_line)
unique_tpl = list(dist_tpl)
print(unique_tpl)
print(max(unique_tpl), min(unique_tpl), sum(unique_tpl)/len(unique_tpl), len(unique_tpl), len(tokens_per_line))
```

```python
minimum = min(unique_tpl)
maximum = max(unique_tpl)
for bucket in range(1, 50):
    buckets = [0 for i in range(bucket)]
    for length in unique_tpl:
        index = int(((length - minimum) / (maximum - minimum) * bucket) // 1)
        buckets[min(index, bucket-1)] += 1
    print(bucket, [float(x) for x in np.quantile(buckets, [0.25, 0.5, 0.75, 1.0])], buckets)
```
